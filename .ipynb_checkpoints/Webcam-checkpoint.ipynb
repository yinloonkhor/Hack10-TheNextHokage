```python
from ultralytics import YOLO
import cv2
import math
import numpy as np


model = YOLO('best.pt')

# start webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

while True:
    success, img = cap.read()
    
    # vertical flip image
    flipped = img[...,::-1,:]
    
    results = model(flipped, stream=True, verbose=False)

    for result in results:
        img = result.plot()
        
    cv2.imshow('Webcam', img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# source: https://github.com/WongKinYiu/yolov7/blob/main/tools/instance.ipynb

## import library (some may be redundant)
import matplotlib.pyplot as plt
import torch
import cv2
import yaml
from torchvision import transforms
import numpy as np

from utils.datasets import letterbox
from utils.general import non_max_suppression_mask_conf

from detectron2.modeling.poolers import ROIPooler
from detectron2.structures import Boxes
from detectron2.utils.memory import retry_if_cuda_oom
from detectron2.layers import paste_masks_in_image

# Intel Realsense Depth Camera D435 is used here
import pyrealsense2
from realsense.realsense_depth import *

import time
import sys
from statistics import median
from math import *
from decimal import Decimal, ROUND_HALF_UP

# For voice output
import pyttsx3

# For speech recognition
from Speech.vosk import Vosk

# for debugging only
from icecream import ic

# setup keyboard interrupt for text output
import signal
def describe_image(signal, frame):
    found = 0
    obj_interest = ask_user(names, vo)
    if obj_interest != "":
        speech = ''
        # describe 3 object nearby in single direction (total 6 direction here )
        clarify = 3
        for i in range(len(detected_obj)):
            if obj_interest == detected_obj[i][0] and obj_interest != "":
                lateral = []; updown  = []; backforth = []; left_object = []; right_object = []; up_object = []; down_object = []; forth_object = []; back_object = []
                found += 1
                x,y,z = detected_obj[i][1:4]

                # Describe the coordinate in more understandable way
                if y > 0: ypos = "top"                
                elif y < 0: ypos = "bottom"                
                else: ypos = "middle"
                    
                if x > 0: xpos = "right"                
                elif x < 0: xpos = "left"               
                else: xpos = "center"            

                # if z == 0
                if z <= 0 or z == 0:
                    position = f"\nIts distance cannot be detected, might be too far/near" if found <= 1 else f"\nThe other one its distance cannot be detected, might be too far/near"

                # if object is near (within 600mm)
                elif z <= 600 and z >0:
                    pos = "near to the camera"
                    if abs(y) <=80 and abs(x) <= 100:
                        position = f"\nIt is around center position {pos}" if found <= 1 else f"\nThe other one is around center position {pos}" 
                    elif abs(y) >= 80 and abs(x) <= 100:
                        position = f"\nIt is around {ypos} position {pos}" if found <= 1 else f"\nThe other one is around {ypos} position {pos}" 
                    elif abs(y) <= 80 and abs(x) >= 100:
                        position = f"\nIt is around {xpos} hand side {pos}" if found <= 1 else f"\nThe other one is around {xpos} hand side {pos}"
                    else:
                        position = f"\nIt is around {ypos} {xpos} position {pos}" if found <= 1 else f"\nThe other one is around {ypos} {xpos} position {pos}"
                
                # if object is at moderate distance
                elif z >600 and z <=5000:
                    pos = "of the camera"
                    ypos = ypos if abs(y) >= 150 else "in front"
                    zsteps = cal_steps(z); xsteps = cal_steps(x)
                    position = f"\nIt is around {zsteps} steps to the front, {xsteps} steps to the {xpos}, and is somewhere {ypos} {pos}, mind your steps" if found <= 1 \
                            else f"\nThe other one is around {zsteps} steps to the front, {xsteps} steps to the {xpos}, and is somewhere {ypos} {pos}, mind your steps"
                
                # far distance object
                else:
                    xpos = xpos if abs(x) >= 150 else ""
                    position = f"\nIt is quite far from the camera, please move slightly forward {xpos} and try again, mind your steps" if found <= 1 \
                            else f"\nThe other one is quite far from the camera, please move slightly forward {xpos} and try again, mind your steps"

                speech += f'{position}.\n' 
                
                if z <= 600:
                    for j in range(len(detected_obj)):                    
                        if detected_obj[j][1:4] != [x,y,z]:     
                            # if dist([detected_obj[j][2]], [y]) < 150 and dist([detected_obj[j][3]], [z]) < 150:# and dist([detected_obj[j][1]], [x]) < 450:
                            #     lateral.append([detected_obj[j][1], detected_obj[j][0]])

                            # if dist([detected_obj[j][1]], [x]) < 150 and dist([detected_obj[j][3]], [z]) < 150:# and dist([detected_obj[j][2]], [y]) < 450:
                            #     updown.append([detected_obj[j][2], detected_obj[j][0]])
                            
                            # if dist([detected_obj[j][1]], [x]) < 150 and dist([detected_obj[j][2]], [y]) < 150 and dist([detected_obj[j][3]], [z]) < 300:
                            #     backforth.append([detected_obj[j][3], detected_obj[j][0]])          
                            xyz = [dist([detected_obj[j][1]], [x]), dist([detected_obj[j][2]], [y]), dist([detected_obj[j][3]], [z])]
                            max_index = [index for index, item in enumerate(xyz) if item == max(xyz)][0]

                            if max_index == 0: lateral.append([detected_obj[j][1], detected_obj[j][0]])
                            elif max_index == 1: updown.append([detected_obj[j][2], detected_obj[j][0]])
                            elif max_index == 2: backforth.append([detected_obj[j][3], detected_obj[j][0]])

                        
                elif z >600 and z <=5000:
                    for j in range(len(detected_obj)):                    
                        if detected_obj[j][1:4] != [x,y,z]:
                            # if dist([detected_obj[j][1]], [x]) < 150 and dist([detected_obj[j][2]], [y]) < 150 and dist([detected_obj[j][3]], [z]) < 450:
                            #     backforth.append([detected_obj[j][3], detected_obj[j][0]])

                            # elif dist([detected_obj[j][2]], [y]) < 200 and dist([detected_obj[j][3]], [z]) < 150 and dist([detected_obj[j][1]], [x]) < 500:
                            #     lateral.append([detected_obj[j][1], detected_obj[j][0]])

                            # elif dist([detected_obj[j][1]], [x]) < 200 and dist([detected_obj[j][3]], [z]) < 150 and dist([detected_obj[j][2]], [y]) < 500:
                            #     updown.append([detected_obj[j][2], detected_obj[j][0]])
                            if dist([detected_obj[j][1]], [x]) < 500 and dist([detected_obj[j][2]], [y]) < 500 and dist([detected_obj[j][3]], [z]) < 450:
                                xyz = [dist([detected_obj[j][1]], [x]), dist([detected_obj[j][2]], [y]), dist([detected_obj[j][3]], [z])]
                                max_index = [index for index, item in enumerate(xyz) if item == max(xyz)][0]

                                if max_index == 0: lateral.append([detected_obj[j][1], detected_obj[j][0]])
                                elif max_index == 1: updown.append([detected_obj[j][2], detected_obj[j][0]])
                                elif max_index == 2: backforth.append([detected_obj[j][3], detected_obj[j][0]])

                lateral.append([x, detected_obj[i][0]])
                lateral.sort()
                lat_pos = -1
                if len(lateral) > 1:
                    speech += f'\nThere are {len(lateral)-1} item(s) beside it.\n'
                    for k in range(len(lateral)):
                        if x == lateral[k][0]:
                            lat_pos = k
                            speech += f'It is {k+1}th item from the left.\n'
                            speech += f'{len(lateral)- k}th item from the right.\n'

                updown.append([y, detected_obj[i][0]])
                updown.sort()
                ud_pos = -1
                if len(updown) > 1:
                    speech += f'\nThere are {len(updown)-1} item(s) above/below of it.\n'
                    for k in range(len(updown)):
                        if y == updown[k][0]:
                            ud_pos = k
                            speech += f'It is {k+1}th item from the bottom.\n'
                            speech += f'{len(updown)- k}th item from the top.\n'

                backforth.append([z, detected_obj[i][0]])
                backforth.sort()
                bf_pos = -1
                if len(backforth) > 1:
                    speech += f'\nThere are {len(backforth)-1} items in backforth of it.\n'
                    for k in range(len(backforth)):
                        if z == backforth[k][0]:
                            bf_pos = k
                            speech += f'It is {k+1}th item from front.\n'
                            speech += f'{len(backforth)- k}th item from behind.\n'

                
                if lat_pos != -1:
                    if lat_pos > 0:
                        try:
                            for a in range(clarify):
                                if lat_pos-(a+1) >=0 : left_object.append(lateral[lat_pos-(a+1)][1]) 
                        except IndexError:
                            pass        
                        speech += f"\nTo the LEFT you will see: "
                        for item in left_object:
                            speech += f"{item}, "

                    if lat_pos < len(lateral)-1:
                        try:
                            for a in range(clarify):
                                right_object.append(lateral[lat_pos+(a+1)][1])
                        except IndexError:
                            pass
                        speech += f"\nTo the RIGHT you will see: "
                        for item in right_object:
                            speech += f"{item}, "

                if ud_pos != -1:
                    if ud_pos > 0:
                        try:
                            for a in range(clarify):
                                if ud_pos-(a+1) >=0 : down_object.append(updown[ud_pos-(a+1)][1])
                        except IndexError:
                            pass
                        speech += f"\nTo the BOTTOM you will see: "
                        for item in down_object:
                            speech += f"{item}, "

                    if ud_pos < len(updown)-1:
                        try:
                            for a in range(clarify):
                                up_object.append(updown[ud_pos+(a+1)][1])
                        except IndexError:
                            pass                    
                        speech += f"\nTo the TOP you will see: "
                        for item in up_object:
                            speech += f"{item}, "

                if bf_pos != -1:
                    if bf_pos > 0:
                        try:
                            for a in range(clarify):
                                if bf_pos-(a+1) >=0 : forth_object.append(backforth[bf_pos-(a+1)][1])
                        except IndexError:
                            pass                     
                        speech += f"\nTo the FORTH you will see: "
                        for item in forth_object:
                            speech += f"{item}, "

                    if bf_pos < len(backforth)-1:
                        try:
                            for a in range(clarify):
                                back_object.append(backforth[bf_pos+(a+1)][1])
                        except IndexError:
                            pass                    
                        speech += f"\nTo the BACK you will see: "
                        for item in back_object:
                            speech += f"{item}, "    

                speech += f'\n'    

        speech = f'I found {found} {obj_interest} in total.\n{speech}' if found else "I cannot locate it, please try again, or move around a little bit.\n"
        print(speech)
        speak(speech)

'''
Voice output
'''
def speak(speech):
    engine = pyttsx3.init()
    engine.setProperty("rate", 150)
    # voices = engine.getProperty('voices')
    # engine.setProperty('voice', voices[1].id)   #changing index, changes voices. 1 for female
    engine.say(speech)
    engine.runAndWait()

'''
Prompt user for object in interest
'''
def ask_user(names, model):
    cont_listen = True
    while cont_listen:
        print("\nPlease tell object in interest (say 'STOP' to quit)")
        speech = ""
        while speech == "": 
            speech = model.listen()
        print(f"'{speech}'")
        recognized_obj = [name for name in names if name in speech]
        if "stop" in speech: 
            cont_listen=False
        elif len(recognized_obj) > 0:
            print(f"Do you mean '{recognized_obj[0]}'? (reply 'YES' or 'NO')")
            confirm = ""
            while "yes" not in confirm and "no" not in confirm:
                confirm = model.listen()
            print(f"'{confirm}'")
            if "yes" in confirm:
                cont_listen = False
                return recognized_obj[0]
        else:    
            print("Unidentified object, please try again")
    
    return ""

''' 
A simple function to display fps 
'''
def display_fps(img, start_time):    
    end_time = time.time()
    fps = 1/np.round(end_time - start_time, 3)
    cv2.putText(img, f'FPS: {int(fps)}', (20,20), cv2.FONT_HERSHEY_PLAIN, 1.5, (0,0,0), 2)


''' 
Preprocessing 
''' 
def image_process (image):       
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)    
    image = letterbox(image, 640, stride=64, auto=True)[0]
    image = cv2.resize(image, (640,480), interpolation= cv2.INTER_AREA)
    image_ = image.copy()
    image = transforms.ToTensor()(image)
    image = torch.tensor(np.array([image.numpy()]))
    image = image.to(device)
    image = image.half()
    output = model(image)

    inf_out, train_out, attn, mask_iou, bases, sem_output = output['test'], output['bbox_and_cls'], output['attn'], output['mask_iou'], output['bases'], output['sem']
    
    bases = torch.cat([bases, sem_output], dim=1)
    nb, _, height, width = image.shape
    
    pooler_scale = model.pooler_scale
    pooler = ROIPooler(output_size=hyp['mask_resolution'], scales=(pooler_scale,), sampling_ratio=1, pooler_type='ROIAlignV2', canonical_level=2)
    
    output, output_mask, output_mask_score, output_ac, output_ab = non_max_suppression_mask_conf(inf_out, attn, bases, pooler, hyp, conf_thres=0.25, iou_thres=0.65, merge=False, mask_iou=None)
    
    pred, pred_masks = output[0], output_mask[0]
    base = bases[0]

    return image, image_, height, width, pred, pred_masks


'''
Object detection process.
All relevant info were drawn or showed inside the image/frame.
Calculate the depth of object from the camera. 
Return labeled frame and approximate depth of the object.  
'''
def obj_detection(image, depth_frame, height, width, pred, pred_masks, intr):
    detected_obj = []

    # Processing    
    bboxes = Boxes(pred[:, :4])
    original_pred_masks = pred_masks.view(-1, hyp['mask_resolution'], hyp['mask_resolution'])
    pred_masks = retry_if_cuda_oom(paste_masks_in_image)( original_pred_masks, bboxes, (height, width), threshold=0.5)
    pred_masks_np = pred_masks.detach().cpu().numpy()
    pred_cls = pred[:, 5].detach().cpu().numpy()
    pred_conf = pred[:, 4].detach().cpu().numpy()
    nimg = image[0].permute(1, 2, 0) * 255
    nimg = nimg.cpu().numpy().astype(np.uint8)
    nimg = cv2.cvtColor(nimg, cv2.COLOR_RGB2BGR)
    nbboxes = bboxes.tensor.detach().cpu().numpy().astype(int)
    pnimg = nimg.copy()

    # Go through detections one by one 
    for one_mask, bbox, cls, conf in zip(pred_masks_np, nbboxes, pred_cls, pred_conf):
        if conf < threshold:
            continue
        color = colors[int(cls)]

        x, y, z = cal_coordinate(one_mask, depth_frame)
        cv2.circle(pnimg, [x,y], 4, (255, 255, 255), -1)
        # cv2.circle(depth_colormap, [x,y], 4, (255, 255, 255), -1)

        ''' source: https://drive.google.com/drive/folders/1-FxgYAKfAFJofLZS13B6m7ihtzJrZ4Sz 
            from: yolo.py'''
        theta = 0  # assume camera is not tilting and is always upright  
        xtemp = z * (x - intr[0]) / intr[2]
        ytemp = z * (y - intr[1]) / intr[3]
        ztemp = z

        # xtarget = xtemp - 35 #35 is RGB camera module offset from the center of the realsense
        xtarget = xtemp 
        ytarget = -(ztemp*sin(theta) + ytemp*cos(theta))
        ztarget = ztemp*cos(theta) + ytemp*sin(theta)

        coordinates_text =  "(" + str(Decimal(str(xtarget)).quantize(Decimal('0'), rounding=ROUND_HALF_UP)) + \
                            ", " + str(Decimal(str(ytarget)).quantize(Decimal('0'), rounding=ROUND_HALF_UP)) + \
                            ", " + str(Decimal(str(ztarget)).quantize(Decimal('0'), rounding=ROUND_HALF_UP)) + ")"    

        # to detect repeated object (havent utilize yet)
        repeat = 0
        for i in range (len(detected_obj)):
            if names[int(cls)] == detected_obj[i][0]:
                repeat += 1
        
        # Put mask
        pnimg[one_mask] = pnimg[one_mask] * 0.5 + np.array(color, dtype=np.uint8) * 0.5
        # Put bounding box 
        cv2.rectangle(pnimg, (bbox[0], bbox[1]), (bbox[2], bbox[3]), color, 2)
        # cv2.rectangle(depth_colormap, (bbox[0], bbox[1]), (bbox[2], bbox[3]), color, 2)

        # Put label to bounding box
        # label = f'{names[int(cls)]}'
        label = f'{names[int(cls)]}:{str(round(conf,2))} | {coordinates_text}mm'
        # label = f'names[int(cls)] | {int(ztarget)}mm'

        t_size = cv2.getTextSize(label, 0, fontScale=0.5, thickness=1)[0]
        c2 = bbox[0] + t_size[0], bbox[1] - t_size[1] - 3
        cv2.rectangle(pnimg, (bbox[0], bbox[1]), c2, color, -1, cv2.LINE_AA)  # filled
        cv2.putText(pnimg, label, (bbox[0], bbox[1] - 2), 0, 0.5, [255, 255, 255], thickness=1, lineType=cv2.LINE_AA)

        # put label at depth frame
        # cv2.rectangle(depth_colormap, (bbox[0], bbox[1]), c2, color, -1, cv2.LINE_AA)  # filled
        # cv2.putText(depth_colormap, label, (bbox[0], bbox[1] - 2), 0, 0.5, [255, 255, 255], thickness=1, lineType=cv2.LINE_AA)    

        # add_to_list(detected_obj, names[int(cls)], xtarget, ytarget, ztarget)
        detected_obj.append([names[int(cls)], xtarget, ytarget, ztarget, repeat])

    return pnimg, detected_obj

''' 
To calculate coordinate.
There are 2 ways to calculate the depth (idea of mine).
@ Method 1 is to calculate depth for all points of mask, then compute the median/upper quartile,
it is accurate, but very slow because need a lot of computation.
@ Method 2 is to calculate the center point of the mask, then compute the depth,
it is fast, but less accurate.
Then proceed to coordinate.
'''
def cal_coordinate(one_mask, depth_frame):
    count = np.count_nonzero(one_mask) 
    center = np.argwhere(one_mask==1).sum(0)/count

    # @ Method 1
    depth_array = []
    for point in np.argwhere(one_mask):
        depth_array.append(depth_frame[point[0],point[1]])
    # depth = median(depth_array)
    depth = np.percentile(depth_array, 75)

    # # @ Method 2     
    # depth = depth_frame[round(center[0]), round(center[1])]  

    # Get coordinate
    coordinate = [round(center[1]), round(center[0]), depth]       

    return coordinate

'''
Simple function to calculate steps required in a given distance D.
Unit in mm.
Assume one step is around 400mm
'''
def cal_steps(D):
    return (round((abs(D)/400),1))

'''
Close camera and all display windows. Save last frame.
'''
def shut_down():
    # turn off RS cam
    dc.release()

    # close all window
    cv2.destroyAllWindows() 

    # Just to save last frame for reference
    cv2.imwrite("recent/depth.jpg", depth_frame)
    cv2.imwrite("recent/depth_color.jpg", depth_colormap) 
    cv2.imwrite("recent/img.jpg", pnimg) 
    
'''
Draw XY axis
'''
def draw_axis(img, h, w):
    # show origin of frame (for reference only)
    ox = w//2
    oy = h//2
    # cv2.circle(pnimg, [ox, oy], 4, (0, 0, 255)) 
    cv2.line(img, [ox,0], [ox,height], (0, 0, 255), 1)
    cv2.line(img, [0,oy], [width,oy], (0, 0,255), 1)  


## ======================= Start here =======================  

# Just for interface 
sys.stdout.write('\rLoading model ......\n')

# setting up device (GPU or else CPU)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Load model from torch
with open('data/hyp.scratch.mask.yaml') as f:
    hyp = yaml.load(f, Loader=yaml.FullLoader)
weigths = torch.load('yolov7-mask.pt')
model = weigths['model']
model = model.half().to(device)
_ = model.eval()

# Names of classes (total 80 from pretrained model)
names = model.names

# Randomly assign one color for each class
colors = np.random.uniform(0, 255, size=(len(names), 3))
# Threshold value for confidence score
threshold = 0.5

# Initialise Intel Realsense Camera
dc = DepthCamera()

# Initialise speech recognition model
vo = Vosk()

while dc.isOpened():

    # Record start time for fps calculation
    start_time = time.time()

    # Read from camera 
    ret, depth_frame, image, depth_colormap, intr = dc.get_frame()    
    
    # Just in case there is no input
    if not ret:
        print('no input')
        continue

    # Preprocess each frame
    image, image_, height, width, pred, pred_masks = image_process(image)

    # if prediction/detection is none
    if pred is None:
        display_fps(image_, start_time)
        image_ = cv2.cvtColor(image_, cv2.COLOR_BGR2RGB)
        pnimg = image_.copy()
        # draw_axis(pnimg, height, width) 
        cv2.imshow("img", pnimg)
        # cv2.imshow("depth", depth_colormap)

    else:
        # Detecting object and display relevant info   
        pnimg, detected_obj = obj_detection(image, depth_frame, height, width, pred, pred_masks, intr)
        # draw_axis(pnimg, height, width)    
        # show result    
        display_fps(pnimg, start_time)
        cv2.imshow("img", pnimg)
        # cv2.imshow("depth", depth_colormap)

    # wait for 1ms before moving to next frame
    key = cv2.waitKey(1)

    # ESC to exit
    if key==27:
        break

    # keyboard interrupt event
    signal.signal(signal.SIGINT, describe_image)

shut_down()
```
