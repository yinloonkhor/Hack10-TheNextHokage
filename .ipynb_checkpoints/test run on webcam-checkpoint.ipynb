```python
def fireball_jutsu(signs):
    success = ['horse','tiger','snake','ram','monkey','boar','horse','tiger']
    csigns = signs.copy()
    cont = False
    ok   = False
    damage = 0
    
    # first sign
    for i in range(len(csigns)):
        if csigns[i] == success[0]:
            cont = True
            del csigns[:i]
            break
        else:
            cont = False
    
    # secong sign
    if cont:
        cont = False
        for j in range(len(csigns)):
            if csigns[j] == success[1]:
                cont = True
                del csigns[:j]
                break
            else:
                cont = False
    
    # third sign
    if cont:
        cont = False
        for k in range(len(csigns)):
            if csigns[k] == success[2]:
                cont = True
                del csigns[:k]
                break
            else:
                cont = False
    
    # fourth sign
    if cont:
        cont = False
        for l in range(len(csigns)):
            if csigns[l] == success[3]:
                cont = True
                del csigns[:l]
                break
            else:
                cont = False
    
    # fifth sign
    if cont:
        cont = False
        for m in range(len(csigns)):
            if csigns[m] == success[4]:
                cont = True
                del csigns[:m]
                break
            else:
                cont = False
    
    # sixth sign
    if cont:
        cont = False
        for n in range(len(csigns)):
            if csigns[n] == success[5]:
                cont = True
                del csigns[:n]
                break
            else:
                cont = False
    
    # seventh sign
    if cont:
        cont = False
        for o in range(len(csigns)):
            if csigns[o] == success[6]:
                cont = True
                del csigns[:o]
                break
            else:
                cont = False
    
    # eighth sign
    if cont:
        cont = False
        for p in range(len(csigns)):
            if csigns[p] == success[7]:
                ok = True
                damage = 300
                del csigns[:p]
                break
            else:
                cont = False                  
            
    if ok:
        return [True, damage, "C:\\Users\\benjy\\Downloads\\naruto\\animation\\fireball_jutsu_vid.mp4"]
    else:
        return [False, damage, ""]
```

```python
import cv2
import numpy as np
#ffpyplayer for playing audio
from ffpyplayer.player import MediaPlayer
video_path="C:\\Users\\benjy\\Downloads\\naruto\\summoning_jutsu_vid.mp4"
def PlayVideo(video_path):
    video=cv2.VideoCapture(video_path)
    player = MediaPlayer(video_path)
    while True:
        grabbed, frame=video.read()
        audio_frame, val = player.get_frame()
        if not grabbed:
#             print("End of video")
            break
        if cv2.waitKey(28) & 0xFF == ord("q"):
            break
        cv2.imshow("Video", frame)
        if val != 'eof' and audio_frame is not None:
            #audio
            img, t = audio_frame
    video.release()
    cv2.destroyAllWindows()
# PlayVideo(video_path)
```

```python
from ultralytics import YOLO
import cv2
import math
import numpy as np


model = YOLO('best.pt')
# model.predict(source=['ox.jpg','ox2.jpg'], save=True, conf=0.1, iou=0.45, show_labels=True)

# start webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

# threshold
thres = 0.4
# object classes
classNames = ["bird",
              "boar",
              "dog",
              "dragon",
              "hare",
              "horse",
              "monkey",
              "ox",
              "ram",
              "rat",
              "snake",
              "tiger",
              "background"
              ]

hand_signs_detected = []

while True:
    success, img1 = cap.read()
    
    # vertically flip image
    img = cv2.flip(img1,1)
    
    results = model(img, stream=True, verbose=False) #conf=thres

    # coordinates
    for r in results:
        boxes = r.boxes

        for box in boxes:
            # bounding box
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2) # convert to int values

            # put box in cam
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 3)

            # confidence
            confidence = math.ceil((box.conf[0]*100))/100
#             print("Confidence --->",confidence)
            
            # class name
            cls = int(box.cls[0])
#             print("Class name -->", classNames[cls])

            # object details
            org = [x1, y1]
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2

            # display text
            text = f'{classNames[cls]} {confidence}'

            cv2.putText(img, text, org, font, fontScale, color, thickness)
            
            hand_signs_detected.append(classNames[cls])
    
    
    cv2.imshow('Webcam', img)
    
    if len(hand_signs_detected) > 1000:
        hand_signs_detected.pop(0)
    
    
    if summoning_jutsu(hand_signs_detected)[0]:
        file=summoning_jutsu(hand_signs_detected)[2]
        PlayVideo(file)
        print(f'Player dealed {summoning_jutsu(hand_signs_detected)[1]} dmg')
        hand_signs_detected=[]
        
    if fireball_jutsu(hand_signs_detected)[0]:
        file=fireball_jutsu(hand_signs_detected)[2]
        PlayVideo(file)
        print(f'Player dealed {fireball_jutsu(hand_signs_detected)[1]} dmg')
        hand_signs_detected=[]
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
